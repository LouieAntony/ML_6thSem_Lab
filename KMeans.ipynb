```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,plot_confusion_matrix,confusion_matrix,accuracy_score
```

```python
data=pd.read_csv("/content/drive/MyDrive/ML Lab test/Datasets/ColourXY.csv")
```

```python
data.head()
```

```python
X=data[['x','y']]
y=data['color']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
km=KMeans()
```

```python
km.fit(X_train,y_train)
predictions=km.predict(X_test)
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
ko=[]
ac=[]
for k in range(1,20):
  km=KMeans(n_clusters=k)
  km.fit(X_train,y_train)
  predictions=km.predict(X_test)
  ko.append(k)
  ac.append(km.inertia_)
```

```python
plt.plot(ko,ac)
```
