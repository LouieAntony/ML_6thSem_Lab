```python
data=pd.read_csv('/content/drive/MyDrive/ML Lab test/Datasets/Advertising.csv')
```

```python
data.head()
```

```python
X=data.drop(labels=['Unnamed: 0','sales'],axis=1)
y=data['sales'].values.reshape(-1,1)
```

```python
reg=LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
reg.fit(X_train,y_train)
```

```python
reg.coef_[0]
```

```python
reg.intercept_[0]
```

```python
def predict(TV,radio,news):
  y=0.04466512*TV+0.19663006*radio+0.00607439*news+2.758071619217109
  return y
```

```python
ans=predict(35,45,34)
```

```python
predictions=reg.predict(X_test)
```
