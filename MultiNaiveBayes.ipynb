```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import plot_confusion_matrix,confusion_matrix,classification_report,accuracy_score,recall_score,plot_roc_curve,matthews_corrcoef,f1_score,precision_score
```

```python
data=pd.read_csv('/content/drive/MyDrive/ML Lab test/Datasets/iris1.csv')
```

```python
data.head()
```

```python
le=LabelEncoder()
data['EncodeClass']=le.fit_transform(data['Class'])
```

```python
data.head()
```

```python
X=data.drop(['Class','EncodeClass'],axis=1)
y=data['EncodeClass']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(X_train,y_train)
```

```python
predictions=mnb.predict(X_test)
```

```python
plot_confusion_matrix(mnb,X_train,y_train)
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(accuracy_score(y_test,predictions))
```

```python
print(recall_score(y_test,predictions,average="weighted"))
```

```python
matthews_corrcoef(y_test,predictions)
```

```python
f1_score(y_test,predictions,average="weighted")
```

```python
precision_score(y_test,predictions,average="weighted")
```
